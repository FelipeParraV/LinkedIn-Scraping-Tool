```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
```

```python
# Title of search
title = "Analyst"

# Location of search
location = "Chile"

# Number of pages (Max 50)
size = 5
```

```python
# Create a list of job ids
id_list = []

# Loop through pages
for i in range(size):
    list_url = f"https://www.linkedin.com/jobs-guest/jobs/api/seeMoreJobPostings/search?keywords=data%20{title}&location={location}&position=1&pageNum=0&start={i*10}"

    response = requests.get(list_url)

    list_data = response.text
    list_soup = BeautifulSoup(list_data, "html.parser")

    page_jobs = list_soup.find_all("li")

    # Scrape job ids
    for job in page_jobs:
        try:
            base_card_div = job.find("div", {"class": "base-card"})
            job_id = base_card_div.get("data-entity-urn").split(":")[3]
            print(job_id)
            id_list.append(job_id)
        except AttributeError:
            continue
```

```python
# Scrape job details

job_list = []
    
for job_id in id_list:
    job_url = f"https://www.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id}"
    job_response = requests.get(job_url)
    # Prints status code:
    print(job_response.status_code)
    # 200 = success
    # 429 = too many requests (blank)
    job_soup = BeautifulSoup(job_response.text, "html.parser")
    job_post = {}
    criteria = job_soup.find_all("span", {"class": "description__job-criteria-text description__job-criteria-text--criteria"})
    
    #Job Tile
    try:
        job_post["job_title"] = job_soup.find("h2", {"class": "top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title"}).text.strip()
    except:
        job_post["job_title"] = None
        
    #Company Name
    try:
        job_post["company_name"] = job_soup.find("a", {"class": "topcard__org-name-link topcard__flavor--black-link"}).text.strip()
    except:
        job_post["company_name"] = None
    
    #Job Criteria (Experience Level, Type of Contract, Job Role, Field)
    try:
        criteria = job_soup.find_all("span", {"class": "description__job-criteria-text description__job-criteria-text--criteria"})
    except:
        criteria = []
    try:
        job_post["experience_level"] = criteria[0].text.strip()
    except:
        job_post["experience_level"] = None
    try:
        job_post["type_of_contract"] = criteria[1].text.strip()
    except:
        job_post["type_of_contract"] = None
    # try:
    #     job_post["job_role"] = criteria[2].text.strip()
    # except:
    #     job_post["job_role"] = None
    # try:
    #     job_post["field"] = criteria[3].text.strip()
    # except:
    #     job_post["field"] = None

    # If job is easy apply
    try:
        if job_soup.find("icon", {"data-svg-class-name": "apply-button__offsite-apply-icon-svg"}):
            job_post["easy_apply"] = False
        else:
            job_post["easy_apply"] = True
    except:
        job_post["easy_apply"] = None

    # Time posted
    try:
        job_post["time_posted"] = job_soup.find("span", {"class": "posted-time-ago__text topcard__flavor--metadata"}).text.strip()
    except:
        job_post["time_posted"] = None

    # Number of applicants
    try:
        job_post["num_applicants"] = job_soup.find("span", {"class": "num-applicants__caption"}).text.strip()
    except:
        job_post["num_applicants"] = None

    # Job Link
    try:
        job_post["job_link"] = f"https://www.linkedin.com/jobs/view/{job_id}/"
    except:
        job_post["job_link"] = None
    
    
    job_list.append(job_post)
```

```python
# Create dataframe
jobs_df = pd.DataFrame(job_list)
jobs_df
```

```python
# Save to csv
jobs_df.to_csv("jobs.csv", index=False)
```
